# 1.- IMPORTAR PANDAS

```python
import pandas as pd
```

# 2.- LEER EL ARCHIVO JSON CON READ_JSON

```python
datos = pd.read_json('/content/dados_vendas_clientes.json')
```

# 3.- APLICAR JSON_NORMALIZE EN LA COLUMNA DADOS_VENDAS

```python
datos = pd.json_normalize(datos['dados_vendas'])
```

# 4.- MOSTRAR VALORES

```python
datos
```

## 5.- COLECTAR LOS VALORES DE LAS COLUMNAS Y VERIFICAR

```python
columnas = list(datos.columns)
columnas
```

# DESTRINCAR LAS LISTAS CON EXPLODE

```python
datos = datos.explode(columnas[1:])
```

# RESETEAR LOS INDEX DE LAS LINEAS

```python
datos.reset_index(drop=True, inplace=True)
```

# OBSERVAR EL DATAFRAME

```python
datos
```

# VERIFICAR LOS TIPOS DE DATOS CON INFO

```python
  datos.info()
```

# LA COLUMNA NUMERICA ES EL 'VALOR DA COMPRA'

```python
datos['Valor da compra']
```

# **INICIAR LA TRANSFORMACION**

# IMPORT DE LA BIBLIOTECA NUMPY

```python
import numpy as np
```

# REMOVER LOS TEXTOS PRESENTES EN LA BASE
# CAMBIAR LAS COMAS SEPARADORAS DEL DECIMAL POR PUNTO

```python
datos['Valor da compra'] = datos['Valor da compra'].apply(lambda x: x.replace('R$', '').replace('.', '').replace(',', '.').strip())
```

# CAMBIAR LOS TIPOS DE DATOS PARA FLOAT

```python
datos['Valor da compra'] = datos['Valor da compra'].astype(np.float64)
```
